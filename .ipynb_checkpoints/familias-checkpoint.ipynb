```python
import json, urllib
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 500)
%matplotlib inline
import scipy.stats as ss
import numpy as np
import geocoder
lista_ciudades=[]
import warnings
# from pandas.core.common import SettingWithCopyWarning
# warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
```

```python
# Trae las 3000+ que si viajan
fam=pd.read_csv('data_familias.csv')
```

```python
# Lee el listado de codigos divipola, selecciona solo codigo y mpio, renombra las columna
divipola=pd.read_excel('Listado_2016.xls',skiprows=1)
d1=divipola[['Código Municipio','Nombre Municipio']]
d1.columns=['mpio_dest','nombre_mpio_dest']

#Se cambia el tipo de dato de int a folat para que en el JOIN con el df de familias sea del mismo tipo
d1['mpio_dest'] = d1['mpio_dest'].astype(float)

#Se eliminan filas duplicadas
d1=d1.drop_duplicates()

#Se hace el join entre familias y divipola para traer el nombre de mpio destino
fam_dest=pd.merge(fam,d1, on='mpio_dest', how='left')

# Se filtran los viajes internos
fam_dest=fam_dest[(fam_dest['viajo']=='si')&(fam_dest['interno']=='si')]

###Para efectos de hacer el mapa filtramos solo los municipos que si encontró. AL parecer hay 2 cod divipola que no encontró.
fam_dest_ok=fam_dest[pd.notnull(fam_dest["nombre_mpio_dest"])]
```

```python
count = 0
def localizar_ciudades(x): 
    city=x['ciudad_origen']
    lista=[tupla for tupla in lista_ciudades if tupla[0]==city]
    if len(lista) < 1:
        try:
            g = geocoder.osm(city+ ' Colombia')   
            lista_ciudades.append((city,g.osm['x'],g.osm['y']))   
            return str(g.osm['x'])+','+str(g.osm['y'])
        except Exception:
            return ','
    else:
        return str(lista[0][1])+','+str(lista[0][2])

def localizar_ciudades_dest(x):
    city=x['nombre_mpio_dest']+' '+x['depto_dest']
    
    lista=[tupla for tupla in lista_ciudades if tupla[0]==x['nombre_mpio_dest']]
    if len(lista) < 1:
        try:            
            count = count + 1
            g = geocoder.osm(city) 
            print('a')
            if g.osm['x'] < 13.5 and g.osm['x'] > -4.2 and g.osm['y'] > -81.9 and g.osm['y'] < -66.8:
                lista_ciudades.append((x['nombre_mpio_dest'],g.osm['x'],g.osm['y']))
                return str(g.osm['x'])+','+str(g.osm['y'])
        except Exception:
            print('b')
            return ','
    else:
        return str(lista[0][1])+','+str(lista[0][2])
```

```python
count = 0
lista_csv=pd.read_csv('coord_ciudades.csv')
lista_csv=lista_csv[['ciudad','lat','lon']]
lista_ciudades = [tuple(x) for x in lista_csv.values]

fam_dest_ok['coord_origen'] = fam_dest_ok.apply(lambda row: localizar_ciudades(row), axis=1)
fam_dest_ok['lat_origen'], fam_dest_ok['lon_origen'] = fam_dest_ok['coord_origen'].str.split(',', 1).str
fam_dest_ok
```

```python
fam_dest_ok['coord_dest'] = fam_dest_ok.apply(lambda row: localizar_ciudades_dest(row), axis=1)
# fam_dest_ok['lat_dest'], fam_dest_ok['lon_dest'] = fam_dest_ok['coord_dest'].str.split(',', 1).str
fam_dest_ok
```
